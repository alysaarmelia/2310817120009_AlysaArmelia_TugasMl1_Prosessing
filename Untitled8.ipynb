Cleaning Data

```python
import cv2
import os
import numpy as np
import hashlib

base_path = '/content/drive/MyDrive/'
folders = ['Tablet (1)', 'Kapsul']
output_path = '/content/drive/MyDrive/DATASET_SIAP_PAKAI/'

target_size = (224, 224)
threshold_blur = 0
min_brightness = 10
max_brightness = 230

if not os.path.exists(output_path):
    os.makedirs(output_path)

hashes_diketahui = set()

laporan = {f: {'awal': 0, 'bersih': 0, 'skip_blur': 0, 'skip_cahaya': 0, 'skip_duplikat': 0} for f in folders}

for folder in folders:
    path_sekarang = os.path.join(base_path, folder)
    path_tujuan = os.path.join(output_path, folder)

    if not os.path.exists(path_tujuan):
        os.makedirs(path_tujuan)

    print(f"\n--- Membersihkan Folder: {folder} ---")

    for filename in os.listdir(path_sekarang):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            laporan[folder]['awal'] += 1
            img_path = os.path.join(path_sekarang, filename)
            img = cv2.imread(img_path)

            if img is None: continue

            img_hash = hashlib.md5(img).hexdigest()
            if img_hash in hashes_diketahui:
                laporan[folder]['skip_duplikat'] += 1
                continue
            hashes_diketahui.add(img_hash)

            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

            skor_blur = cv2.Laplacian(gray, cv2.CV_64F).var()

            rata_cahaya = np.mean(gray)

            if skor_blur < threshold_blur:
                laporan[folder]['skip_blur'] += 1
                continue

            if rata_cahaya < min_brightness or rata_cahaya > max_brightness:
                laporan[folder]['skip_cahaya'] += 1
                continue

            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            img_resized = cv2.resize(img_rgb, target_size, interpolation=cv2.INTER_AREA)

            laporan[folder]['bersih'] += 1
            save_path = os.path.join(path_tujuan, f"clean_{filename}")
            cv2.imwrite(save_path, cv2.cvtColor(img_resized, cv2.COLOR_RGB2BGR))

print("\n" + "="*30)
print("LAPORAN CLEANING DATASET")
print("="*30)
for f, stat in laporan.items():
    print(f"Folder: {f}")
    print(f"  - Total Awal      : {stat['awal']}")
    print(f"  - Dibuang (Blur)  : {stat['skip_blur']}")
    print(f"  - Dibuang (Cahaya): {stat['skip_cahaya']}")
    print(f"  - Dibuang (Duplikat): {stat['skip_duplikat']}")
    print(f"  - Hasil Bersih    : {stat['bersih']}")
    print("-" * 20)
```
